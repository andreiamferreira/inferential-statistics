## Correlação de Pearson

````python
# Bibliotecas utilizadas
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
````

````python
from google.colab import drive
drive.mount('/content/gdrive')
````

````python
df_respostas = pd.read_excel('/content/gdrive/MyDrive/Forms/respostas.xlsx', sheet_name='Respostasaoform')

df_respostas.head(14)
````

````python
# Separação dos grupos: listagem das colunas utilizadas determinarmos X e Y
idade = df_respostas['Idade'].tolist()
peso = df_respostas['Peso'].tolist()

#print(idade)
#print(peso)

print("DataFrame para melhor visualização")
df_dados = pd.DataFrame({
    'Idade (X)': idade,
    'Peso (Y)': peso
})

print(df_dados)
````

````python
# Correlação Pearson para idade (X) x peso (Y)
# Função pearsonr determina o coeficiente de Pearson de x(idade) em
# relação à y(peso)
coeficiente_pearson, p_value = pearsonr(idade, peso)

print(coeficiente_pearson)
````

**Análise**:

O coeficiente de correlação de Pearson permite fazer uma estimativa da correlação entre as variáveis X e Y, que neste caso são a idade e o peso, respectivamente. Considerando que uma correlação linear direta é definida como 0 < coeficiente de Pearson < 1, e o nosso coef. Pearson foi de 0.3035, podemos assumir que existe uma correlação linear direta entre a idade e o peso dos alunos, que pode ser demonstrado no Diagrama de Dispersão abaixo.

````python
# Diagrama de Dispersão
plt.scatter(idade, peso)
plt.title("Diagrama de Dispersão")
plb.xlabel("Idade (anos)")
plb.ylabel("Peso (Kg)")
plt.show()
````

### Regresão Linear Simples

Para a regressão linear simples, utilizamos a biblioteca statsmodels, que possui a função OLS (Ordinary Least Squares, o chamado Método dos Mínimos Quadrados), para descrever a relação entre as variáveis de idade e peso através de uma função. No Método dos Mínimos Quadrados a soma dos desvios dos pontos em relação à reta é 0, e a soma dos quadrados dos desvios deve ser mínima - na representação gráfica, isso permite a reta da função linear seja traçada nos pontos do modelo (e isso é obtido com o método *fit()*).



Abaixo, o código e os resultados da regressão linear. Antes de analisarmos os resultados, é importante ressaltar que utilizamos a importação explícita do pacote api (*statsmodels.api*) pois a importação direta não funcionou.


Ademais, definimos que no nosso modelo a idade seria X e o peso Y (assim como definido anteriormente), sendo o código original:


```
   res = sm.OLS(y, X).fit()

```

````python
import statsmodels.api as sm

modelo = sm.OLS(peso, idade).fit()
print(modelo.summary())
````

**Análise:**

*Coeficiente de correlação (R):* 0,30.

*Coeficiente de determinação:* 0,94 (ou 0,94%).


Assim, sendo R = 0,305, Xi (idade) e Yi (peso) estão correlacionadas,  indicando uma correlação direta - quanto maior a idade, maior o peso. Já o coeficiente de determinação estima que 94% da variabilidade da idade é explicada pela variabilidade do peso.





````python
plb.figure(figsize=(8, 4))
plb.scatter(idade, peso)
plb.plot(idade, modelo.predict(), color="r")
plb.title("Idade x Peso")
plb.xlabel("Idade (anos)")
plb.ylabel("Peso (Kg)")
````
