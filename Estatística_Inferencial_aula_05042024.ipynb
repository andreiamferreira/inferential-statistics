# **Em relação as variáveis quantitativas, será calculado a Média, Moda, Mediana, Desvio Padrão, Amplitude e Variação de Coeficiente.**

##**Cálculo das Médias**

```python
#Média de  idade, peso, ano, Frequênciatvfisica

m_idade = df_respostas["Idade"].mean()
m_peso = df_respostas["Peso"].mean()
m_ano = df_respostas["Ano"].mean()
m_frequencia = campo_vazio3['Frequenciatvfisica'].mean()

print(f"A média da idade é: {m_idade:.2f}")
print(f"A média do peso é: {m_peso:.2f}")
print(f"A média do ano do curso é: {m_ano:.2f}")
print(f"A média da freqência que pratica atividade física é: {m_frequencia:.2f}")
```

## **Cálculo das Medianas**

```python
med_idade = df_respostas["Idade"].median()
med_peso = df_respostas["Peso"].median()
med_ano = df_respostas['Ano'].median()
med_frequencia = campo_vazio3['Frequenciatvfisica'].median()

print(f"A mediana da idade é: {med_idade:.2f}")
print(f"A mediana do peso é: {med_peso:.2f}")
print(f"A mediana do ano do curso é: {med_ano:.2f}")
print(f"A mediana da frequência que pratica atividade física é: {med_frequencia:.2f}")
```

## **Cálculo da Moda**

```python
md_idade = df_respostas['Idade'].mode().tolist()
md_peso = df_respostas['Peso'].mode().tolist()
md_ano = df_respostas['Ano'].mode().tolist()
md_frequencia = campo_vazio3['Frequenciatvfisica'].mode().tolist()


print(f"A moda da idade é: {''.join(map(str, md_idade))}")
print(f"A moda do peso é: {''.join(map(str, md_peso))}")
print(f"A moda do ano é: {''.join(map(str, md_ano))}")
print(f"A moda da frequência que pratica atividade física é: {''.join(map(str, md_frequencia))}")
```

## **Desvio padrão**

```python
dp_idade = st.stdev(df_respostas['Idade'])
dp_peso = st.stdev(df_respostas['Peso'])
dp_ano = st.stdev(df_respostas['Ano'])
dp_frequencia = (campo_vazio3['Frequenciatvfisica']).std()


print(f"O desvio padrão da idadde: {dp_idade:.2f}")
print(f"O desvio padrão do peso: {dp_peso:.2f}")
print(f"O desvio padrão do ano do curso: {dp_ano:.2f}")
print(f"O desvio padrão da frequência que pratica atividade física: {dp_frequencia:.2f}")

```

## **Amplitude**

```python
amp_idade = df_respostas['Idade'].max() - df_respostas['Idade'].min()
amp_peso = df_respostas['Peso'].max() - df_respostas['Peso'].min()
amp_ano = df_respostas['Ano'].max() - df_respostas['Ano'].min()
amp_frequencia = campo_vazio3['Frequenciatvfisica'].max() - campo_vazio3['Frequenciatvfisica'].min()

print(f"A amplitude da idadde é: {amp_idade:.2f}")
print(f"A amplitude do peso é: {amp_peso:.2f}")
print(f"A amplitude do ano do curso é: {amp_ano:.2f}")
print(f"A amplitude da frequência que pratica atividade física é: {amp_frequencia:.2f}")
```

## **Coeficiente de Variação**

```python
cof_idade = st.variance(df_respostas["Idade"]) / m_idade
cof_peso = st.variance(df_respostas["Idade"]) / m_peso
cof_ano = st.variance(df_respostas["Ano"]) / m_ano
cof_frequencia = (campo_vazio3["Frequenciatvfisica"]).var()

print(f"O coeficiente de variação da idade é: {cof_idade:.2f},")
print(f"O coeficiente de variação do peso é: {cof_peso:.2f},")
print(f"O coeficiente de variação do ano do curso é: {cof_ano:.2f}.")
print(f"O coeficiente de variação da frequência que pratica atividade física é: {cof_frequencia:.2f}.")
```
