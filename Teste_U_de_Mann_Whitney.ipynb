# **TESTES NÃO-PARAMÉTRICOS: Teste U de Mann-Whitney**

Variáveis comparadas: ano de curso (qualitativa ordinal) e pratica de atividade física. Para formular as hipóteses, limitamos o ano do curso aos dois primeiros.

> H0: A quantidade de alunos que estão no ciclo básico do curso (primeiro e segundo ano) é igual entre alunos que praticam atividades físicas e os que não praticam.

> H1: A quantidade de alunos que estão no ciclo básico do curso (primeiro e segundo ano) é diferente entre alunos que praticam atividades físicas e os que não praticam.



```python
import pandas as pd
import statistics as st
import numpy as np
import statsmodels.stats.weightstats as ssw
import scipy.stats as stats
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_respostas = pd.read_excel('/content/gdrive/MyDrive/Forms/respostas.xlsx', sheet_name='Respostasaoform')

df_respostas.head(14)
```

```python
# Separação dos grupos: alunos do primeiro e segundo ano que praticam atividade física e alunos que não praticam
alunos_ativos_ciclo_basico = df_respostas[(df_respostas['Ano'].isin([1, 2])) & (df_respostas['AtivFisica'] == 'Sim')]
grupo_alunos_1 = alunos_ativos_ciclo_basico['Ano'].tolist()

alunos_inativos_ciclo_basico = df_respostas[(df_respostas['Ano'].isin([1, 2])) & (df_respostas['AtivFisica'] == 'Não')]
grupo_alunos_2 = alunos_inativos_ciclo_basico['Ano'].tolist()

print(grupo_alunos_1)
print(grupo_alunos_2)
```

```python
stats.mannwhitneyu(grupo_alunos_1, grupo_alunos_2)
```

Considerando o nível de significância de 5%, o p-value = 0.6708 não é menor do que 0.05, portanto, não há evidência o suficiente para rejeitar a hipótese nula.

```python
# PROVA REAL: Somando os rankings:
# n de w grupo 1
print('na = ' + str(len(grupo_alunos_1)))
print('nb = ' + str(len(grupo_alunos_2)))

soma_ano_grupo_1 = alunos_ativos_ciclo_basico['Ano'].sum()
soma_ano_grupo_2 = alunos_inativos_ciclo_basico['Ano'].sum()

print('Soma  do grupo 1: ' + str(soma_ano_grupo_1))
print('Soma do grupo 2: ' + str(soma_ano_grupo_2))
```

Considerando o nível de significância de 5%, o p-value de 0.6708, a quantidade amostral, o somatório dos rankings e a tabela disponível no material, não é possível concluir a rejeição da hipótese nula, pois não há valores na tabela para essa quantidade amostral - portanto, não há evidências para rejeitar H0.
